# Generate Random City Data

```python
output_data_file = "output/cities.csv"

lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
lat_lngs = []
cities = []

lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)


for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    if city not in cities:
        cities.append(city)

len(cities)
```

```python
cities_df["Cloudiness"]=""
cities_df["Country"]=""
cities_df["Date"]=""
cities_df["Humidity"]=""
cities_df["Lat"]=""
cities_df["Lng"]=""
cities_df["Max Temp"]=""
cities_df["Wind Speed"]=""
```

```python
for index, row in cities_df.iterrows(): 
    target_city=row["city"]
    query_url= f"{base_url}appid={api_key}&q={target_city}"
    response=requests.get(query_url).json()
    try: 
        print(response["clouds"]["all"])
    except KeyError: 
        print("NA")
```

```python
x=0
for index, row in cities_df.iterrows(): 
    target_city=row["city"]
    query_url= f"{base_url}appid={api_key}&q={target_city}"
    response=requests.get(query_url).json()
    try:
        clouds=response["clouds"]["all"]
        cities_df.set_value(x, "Cloudiness", clouds)
    except KeyError:
        cities_df.set_value(x, "Cloudiness", "NaN")
    try:
        country=response["sys"]["country"]
        cities_df.set_value(x, "Cloudiness", country)
    except KeyError:
        cities_df.set_value(x, "Cloudiness", "NaN")
    try:
        date=response["dt"]
        cities_df.set_value(x, "Date", date)
    except KeyError:
        cities_df.set_value(x, "Date", "NaN")
    try:
        humidity=response["main"]["humidity"]
        cities_df.set_values(x, "Humidity", humidity)
    except KeyError:
        cities_df.set_value(x, "Humidity", "NaN")
    try:    
        lat=response["coord"]["lat"]
        cities_df.set_value(x, "Lat", lat)
    except KeyError:
        cities_df.set_value(x, "Lat", "NaN")
    try:
        lng=response["coord"]["lon"]
        cities_df.set_value(x, "Lng", lng)
    except KeyError:
        cities_df.set_value(x, "Lng", "NaN")
    try:
        max_temp=response["main"]["temp_max"]
        cities_df.set_value(x, "Max Temp", max_temp)
    except KeyError:
        cities_df.set_value(x, "Max Temp", "NaN")
    try:
        wind=response["wind"]["speed"]
        cities_df.set_value(x, "Wind Speed", wind)
    except KeyError: 
        cities_df.set_value(x, "Wind Speed", "NaN")
            
    x+=1
    
```
