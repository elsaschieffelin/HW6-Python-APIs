```python
#import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import requests
import time
from time import gmtime
import json
from pprint import pprint
from config import api_key
from citipy import citipy
```

# Generate Random City Data

```python
#create output file
output_data_file = "output/cities.csv"
#set lat and lng limits 
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
#create coord list and cities list
lat_lngs = []
cities = []
#pull random lat and lng and create coordinates. store in lat_lngs
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

#loop through list to identify cities, store cities in city list
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    if city not in cities:
        cities.append(city)

len(cities)
```

# API Calls

```python
#test the list of cities to see if the Open Weather API has information on each of them. If not, 
x=0
z=1
y=1
cities_edit=[]
for city in cities:
    base_url = "http://api.openweathermap.org/data/2.5/weather?"
    city = cities[x]
    x+=1
    target_url = f"{base_url}&q={city}&appid={api_key}"
    try:
        response=requests.get(target_url).json()
        response["clouds"]["all"]
        print(f"Processing Record {z} of Set {y}, {city}")
        cities_edit.append(city)
    except KeyError: 
        print("city not found. Skipping...")
    z+=1
    if z==51:
        y+=1
        z=1
```

```python
#see the length of the list of the remaining cities
len(cities_edit)
```

```python
#create a dataframe
cities_df=pd.DataFrame({"City": cities_edit})
```

```python
#add columns to the dataframe where we can store the infomration that we are analyzing
cities_df["Cloudiness"]=""
cities_df["Country"]=""
cities_df["Date"]=""
cities_df["Humidity"]=""
cities_df["Lat"]=""
cities_df["Lng"]=""
cities_df["Max Temp"]=""
cities_df["Wind Speed"]=""
```

```python
#loop through every row in the dataframe, access API and find the information that we will want to store in the empty columns
#i.e. cloudiness, country, date, humidity, lat, lng, max temp, wind speed
#even though we should have informaiton on all of these cities, add try except blockin case we find missing information
x=0
for index, row in cities_df.iterrows(): 
    target_city=row["City"]
    query_url= f"{base_url}appid={api_key}&q={target_city}&units=imperial"
    response=requests.get(query_url).json()
    try:
        clouds=response["clouds"]["all"]
        cities_df.set_value(x, "Cloudiness", clouds)
    except KeyError:
        cities_df.set_value(x, "Cloudiness", np.NAN)
    try:
        country=response["sys"]["country"]
        cities_df.set_value(x, "Country", country)
    except KeyError:
         cities_df.set_value(x, "Country", np.NAN)
    try:
        date=response["dt"]
        cities_df.set_value(x, "Date", date)
    except KeyError:
         cities_df.set_value(x, "Date", np.NAN)
    try:
        humidity=response["main"]["humidity"]
        cities_df.set_value(x, "Humidity", humidity)
    except KeyError:
         cities_df.set_value(x, "Humidity", np.NAN)
    try:    
        lat=response["coord"]["lat"]
        cities_df.set_value(x, "Lat", lat)
    except KeyError:
         cities_df.set_value(x, "Lat", np.NAN)
    try:
        lng=response["coord"]["lon"]
        cities_df.set_value(x, "Lng", lng)
    except KeyError:
         cities_df.set_value(x, "Lng", np.NAN)
    try:
        max_temp=response["main"]["temp_max"]
        cities_df.set_value(x, "Max Temp", max_temp)
    except KeyError:
         cities_df.set_value(x, "Max Temp", np.NAN)
    try:
        wind=response["wind"]["speed"]
        cities_df.set_value(x, "Wind Speed", wind)
    except KeyError: 
         cities_df.set_value(x, "Wind Speed", np.NAN)
            
    x+=1
```

```python
#display dataframe
cities_df = cities_df.dropna()
```

```python
#export dataframe 
cities_df.to_csv(output_data_file)
```
